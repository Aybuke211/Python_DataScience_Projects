**Veri Temizleme Aşaması**

```python
# kütüphaneleri import edelim
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# Pythonda uyarıları kapatalım
import warnings
warnings.filterwarnings("ignore")
```

```python
# veriyi içeri aktaralım
veri = pd.read_csv("/olimpiyatlar.csv")
veri.head()
```

```python
veri.info()
```

```python
veri.columns
```

```python
# Sütun isimlerini değiştirelim
veri.rename(columns={'ID': 'id',
                     'Name': 'isim',
                     'Gender': 'cinsiyet',
                     'Age': 'yas',
                     'Height': 'boy',
                     'Weight': 'kilo',
                     'Team': 'takim',
                     'NOC': 'uok',
                     'Games': 'oyunlar',
                     'Year': 'yil',
                     'Season': 'sezon',
                     'City': 'sehir',
                     'Sport': 'spor',
                     'Event': 'etkinlik',
                     'Medal': 'madalya'}, inplace= True)
veri.head(2)
```

```python
# drop fonksiyonu ile id ve oyunlar sütunlarını çıkaralım
veri = veri.drop(["id","oyunlar"],axis=1)
veri.head(2)
```

```python
# Boy ve Kilo Sütunu Kayıp Veri Doldurma
essiz_etkinlik = pd.unique(veri.etkinlik)
print("Eşsiz etkinlik sayısı: {}".format(len(essiz_etkinlik)))
essiz_etkinlik[:10]
```

```python
# her bir etkinliği iteratif olarak dolaş
# etkinlik özelinde boy ve kilo ortalamalarını hesapla ve bul
# etkinlik özelinde kayıp boy ve kilo değerlerini etkinlik ortalamalarına eşitle
veri_gecici = veri.copy() # gerçek veriyi bozmamak için bir kopyası oluşturuldu.
boy_kilo_liste = ["boy","kilo"]
for e in essiz_etkinlik:
  etkinlik_filtre = veri_gecici.etkinlik == e
  # veriyi etkinliğe göre filtrele
  veri_filtreli = veri_gecici[etkinlik_filtre]
  # boy ve kilo için etkinlik özelinde ortalamaları hesaplayalım
  for s in boy_kilo_liste:
    ortalama = np.round(np.mean(veri_filtreli[s]),2)
    if ~np.isnan(ortalama): # eğer etkinlik özelinde ortalama varsa
      veri_filtreli[s] = veri_filtreli[s].fillna(ortalama)
    else: # eğer etkinlik özelinde ortalama yoksa
      tum_veri_ortalamasi = np.round(np.mean(veri[s]),2)
      veri_filtreli[s] = veri_filtreli[s].fillna(tum_veri_ortalamasi)
  # etkinlik özelinde kayıp değerleri doldurulmuş olan veriyi veri_gecici'ye eşitle
  veri_gecici[etkinlik_filtre] = veri_filtreli
  # kayıp değerleri giderilmiş olan geçici veriyi gerçek veriye eşitleyelim
  veri = veri_gecici.copy()
  veri.info() # boy ve kilo sütunlarında kayıp değer sayısına bak



```

```python
# Yaş Sütunu Kayıp Veri Doldurma
# Yaş sütununda bulunan kayıp veriyi veri setinin yaş ortalamasına göre dolduracağız
# yas değişkeninde tanımlı olmayan değerleri bulalım
yas_ortalamasi = np.round(np.mean(veri.yas),2)
print("Yaş Ortalaması: {}".format(yas_ortalamasi))
veri["yas"] = veri["yas"].fillna(yas_ortalamasi)
veri.info()


```

```python
# Madalya Alamayan Sporcuları Veri Setinden Çıkar
madalya_degiskeni = veri["madalya"]
pd.isnull(madalya_degiskeni).sum()
```

```python
madalya_degiskeni_filtresi = ~pd.isnull(madalya_degiskeni) # madalya almış olanlar
veri = veri[madalya_degiskeni_filtresi] # sadece madalya almış olanlar
veri.head(5)
```

```python
veri.info()
```

```python
# sonradan kullanabilmek için veriyi kaydedelim
veri.to_csv("olimpiyatlar_temizlenmis.csv", index = False)
```

**Veri Analiz Aşaması**

```python

# Öncelikli olarak histogram grafiklerini çizdireceğimiz fonksiyonu yaz
def plotHistogram(degisken):
  plt.figure()
  plt.hist(veri[degisken], bins = 85, color = "pink")
  plt.xlabel(degisken)
  plt.ylabel("Frekans")
  plt.title("Veri Sıklığı - {}".format(degisken))
  plt.show()


```

```python
# Tüm sayısal değişkenler için histogramları çiz
sayısal_degisken = ["yas","boy","kilo","yil"]
for i in sayısal_degisken:
  plotHistogram(i)
```
